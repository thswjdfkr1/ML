```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(X)
input_scaled = ss.transform(X)
print(input_scaled[0:5,])
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(input_scaled)
wine_pca = pca.transform(input_scaled)
print(wine_pca.shape)
print(wine_pca[0:6,])
```

```python
import pandas as pd
wine_pca=pd.DataFrame(wine_pca,columns=['pc1','pc2'])
print(wine_pca)
```

```python
print(pca.explained_variance_ratio_)
import numpy as np
print(np.sum(pca.explained_variance_ratio_))
```

```python
pca_all = PCA(n_components=2)
pca_all.fit(input_scaled)
wine_pca_all = pca_all.transform(input_scaled)
print(wine_pca_all.shape)
print(wine_pca_all[0:6,])
print(pca_all.explained_variance_ratio_)
print(np.sum(pca_all.explained_variance_ratio_))
```
