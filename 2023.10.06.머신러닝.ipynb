```python
import seaborn as sns
iris = sns.load_dataset('iris')
input_var = iris[['sepal_length','sepal_width','petal_length','petal_width']]
target_var = iris['species']
print(input_var)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300,random_state=0)
kmeans.fit(input_var)
print(kmeans.labels_)
print(kmeans.cluster_centers_)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2, init='k-means++', max_iter=300,random_state=0)
kmeans.fit(input_var)
print(kmeans.labels_)
print(kmeans.cluster_centers_)
```

```python
input_var['target'] = target_var
input_var['cluster']=kmeans.labels_
iris_result = input_var.groupby(['target','cluster'])['sepal_length'].count()
print(iris_result)
```

```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
import pandas as pd
iris = load_iris()
print(iris.data[0:5])
print(iris.target[0:5])
```

```python
variable_names = ['sepal_length','sepal_width','petal_length','petal_width']
iris_pd = pd.DataFrame(data=iris.data, columns=variable_names)
print(iris_pd)
```

```python
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300,random_state=1234).fit(iris_pd)
iris_pd['cluster'] = kmeans.labels_
print(iris_pd)
```

```python
score_samples = silhouette_samples(iris.data, iris_pd['cluster'])
print(score_samples.shape)
print(score_samples[0:10])
```

```python
iris_pd['silhouette_coeff'] = score_samples
print(iris_pd)
```

```python
average_score = silhouette_score(iris.data, iris_pd['cluster'])
print('Silhoutte Coefficient={0:.3f}'.format(average_score))
print(iris_pd.groupby('cluster')['silhouette_coeff'].mean())
```
