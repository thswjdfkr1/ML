```python
import pandas as pd
url="https://archive.ics.uci.edu/ml/machine-learning-databases/abalone/abalone.data"
AB=pd.read_csv(url, header=None, names=['Sex','Length','Diameter','Height','Whole','Shucked','Viscera','Shell','Rings'])
print(AB.head())
```

```python
AB_data=AB.drop(['Sex'],axis=1)
print(AB_data)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
X=AB[['Length','Height']]
y=AB['Sex']
train_input, test_input, train_target, test_target = train_test_split(
    X,y, test_size=0.25,random_state=1004)
k_nn = KNeighborsClassifier()
k_nn.fit(train_input, train_target)
print(k_nn.score(train_input, train_target))
print(k_nn.score(test_input, test_target))
```

```python
AB_data=AB.drop(['Sex'],axis=1)
AB_data
```

```python
input_var = AB_data[['Length','Diameter','Height','Whole',
  'Shucked','Viscera','Shell','Rings']]
target_var = AB['Sex']
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(input_var)
input_scaled=ss.transform(input_var)
```

```python
from sklearn.decomposition import PCA
import numpy as np
for k in range(0, 9):
  pca = PCA(n_components = k)
  pca.fit(input_scaled)
  adult_pca = pca.transform(input_scaled)
  if np.sum(np.sum(pca.explained_variance_ratio_)) > 0.95:
    print(adult_pca.shape)
    print(pca.explained_variance_ratio_)
    print("K=",k,":",np.sum(pca.explained_variance_ratio_))
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
for k in range(2,11):
  kmeans = KMeans(n_clusters=k, init='k-means++', max_iter=300,random_state=45982)
  kmeans.fit(input_var)
  #print(kmeans.labels_)
  S=silhouette_score(input_var,kmeans.labels_)
  print("K=",k,":",S)
```
